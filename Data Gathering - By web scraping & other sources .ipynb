```python
# Considering any review with rating 3 or above as positive, and reviews with rating 1 and 2 as negative.
# To make sure the data is balanced, I will consider a varying number of rows of each rating.

df2_5stars = df2[df2['customer_stars'] == 5].sample(6000)
df2_4stars = df2[df2['customer_stars'] == 4].sample(3200)
df2_3stars = df2[df2['customer_stars'] == 3].sample(5500)
df2_2stars = df2[df2['customer_stars'] == 2].sample(12500)
df2_1stars = df2[df2['customer_stars'] == 1].sample(25000)
```
